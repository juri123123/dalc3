```python
import requests
from bs4 import BeautifulSoup

book_review = []
url = 'http://booklog.kyobobook.co.kr/blog/listSubMainReview.do'
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')

#contents > div.list_wrap > div:nth-child(4) > p
#contents > div.list_wrap > div:nth-child(1) > div.title > a
#contents > div.list_wrap > div:nth-child(30) > p
pages = range(1,30)
for page in pages:
  src = '#contents > div.list_wrap > div:nth-of-type({})> p'.format(page)
  src2 = '#contents > div.list_wrap > div:nth-of-type({}) > div.title > a'.format(page)
  reviews = soup.select(src)
  names = soup.select(src2)
  book_review.extend([names,reviews])

book_review
```
